```python
def match_casing(input_text, output_text):
    if input_text[0].islower():
        return output_text[0].lower() + output_text[1:]
    return output_text
```

```python
def query_gpt(prompt, input_str, model="gpt-4o-mini"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.chat.completions.create(model=model, messages=messages, temperature=0.1)

    raw_output = response.choices[0].message.content.strip()

    if not input_str.strip().endswith(".") and raw_output.endswith("."):
        raw_output = raw_output[:-1]

    final_output = match_casing(input_str, raw_output)
    return final_output
```

```python
import evaluate
from tqdm import tqdm

exact_match = evaluate.load("exact_match")
bleu = evaluate.load("bleu")
rouge = evaluate.load("rouge")
import bert_score


def compute_metrics(preds, refs, tokenizer):
    em = exact_match.compute(predictions=preds, references=refs)["exact_match"]
    bleu_score = bleu.compute(predictions=preds, references=refs)["bleu"]
    rouge_score = rouge.compute(predictions=preds, references=refs)
    bertscore = bert_score.score(preds, refs, lang="en", verbose=True)[2].mean().item()

    token_mismatches = 0
    total = len(preds)

    for pred, ref in zip(preds, refs):
        pred_len = len(tokenizer.encode(pred, add_special_tokens=False))
        ref_len = len(tokenizer.encode(ref, add_special_tokens=False))
        if pred_len != ref_len:
            token_mismatches += 1

    token_match_rate = (total - token_mismatches) / total

    return {
        "Exact Match": round(em, 4),
        "BLEU": round(bleu_score, 4),
        "ROUGE-L": round(rouge_score["rougeL"], 4),
        "BERTScore (F1)": round(bertscore, 4),
        "Token Length Match Rate": round(token_match_rate, 4)
    }
```

```python
def generate_swapped_response(input_text, inference_model, tokenizer, device=device):
  inputs = tokenizer(input_text, return_tensors="pt").to(device)

  with torch.no_grad():
      outputs = inference_model.generate(**inputs, max_new_tokens=100, temperature=0.1)

  generated = tokenizer.decode(outputs[0], skip_special_tokens=True)

  return generated
```

```python
import tiktoken

encoding = tiktoken.encoding_for_model("gpt-4")

def gpt_token_count(text):
    return len(encoding.encode(text))
```

```python
def compute_metrics_gpt(preds, refs):
    em = exact_match.compute(predictions=preds, references=refs)["exact_match"]
    bleu_score = bleu.compute(predictions=preds, references=refs)["bleu"]
    rouge_score = rouge.compute(predictions=preds, references=refs)
    bertscore = bert_score.score(preds, refs, lang="en", verbose=True)[2].mean().item()

    token_mismatches = 0
    total = len(preds)

    for pred, ref in zip(preds, refs):
        pred_len = gpt_token_count(pred)
        ref_len = gpt_token_count(ref)
        if pred_len != ref_len:
            token_mismatches += 1

    token_match_rate = (total - token_mismatches) / total

    return {
        "Exact Match": round(em, 4),
        "BLEU": round(bleu_score, 4),
        "ROUGE-L": round(rouge_score["rougeL"], 4),
        "BERTScore (F1)": round(bertscore, 4),
        "Token Length Match Rate": round(token_match_rate, 4)
    }
```
