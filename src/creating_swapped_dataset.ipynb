```python
def find_orig_sent_id(file_name, tag, df):
    txt_file_name = os.path.basename(file_name).replace("_1.ann", ".txt")
    # print(txt_file_name)
    df = df[df['original_file_name']==txt_file_name]
    for x, row in df.iterrows():
        # print(row['changed_ann'])
        swapped_annotations = ast.literal_eval(row['changed_ann'])
        for ind, ent in swapped_annotations.items():
            # print(ent, t)
            if ind == tag:
                return str(row['orig_sent_id'])
            
def process_ann_file(file_path, destination_path, df):
    modified_lines = []

    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            parts = line.strip().split('\t')
            # print("parts:", parts)
            if len(parts) == 5:
                # print("here")
                tag, label, start, end, text = parts
                orig_sent_id = find_orig_sent_id(file_path, tag, df)
                if orig_sent_id:
                    orig_sent_id = int(orig_sent_id)
                    modified_line = f'{tag}\t{label}\t{start}\t{end}\t{text}\t{orig_sent_id}\n'
                    modified_lines.append(modified_line)
                    print("id:", orig_sent_id)
                else:
                    modified_lines.append(line)
            else:
                modified_lines.append(line)
    # print(destination_path)
    with open(destination_path, 'w', encoding='utf-8') as file:
        file.writelines(modified_lines)
```

```python
for source_path in bruk_changed_files_ann:
    destination_dir = '/Users/linndfors/study/diploma/ner_for_fem/data/v2.0-swapped/bruk'

    file_name = os.path.basename(source_path)
    file_name = file_name.replace("_1.ann", "-swapped.ann")

    destination_path = os.path.join(destination_dir, file_name)
    process_ann_file(source_path, destination_path, parallel_dataset_bruk_dataset)
    # print(f'Processed {destination_path}')

    # shutil.copy(source_path, destination_path)
```
