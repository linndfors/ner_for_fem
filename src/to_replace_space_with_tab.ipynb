```python
import os

# Directory containing the .ann files
directory = '/Users/linndfors/study/diploma/ner_for_fem/data/v2.0-swapped_with_tab/ng'

# Loop through all files in the directory
for filename in os.listdir(directory):
    if filename.endswith('-swapped.ann'):
        file_path = os.path.join(directory, filename)
        
        # Open the .ann file for reading
        with open(file_path, 'r', encoding='utf-8') as file:
            lines = file.readlines()

        # Open the same file for writing the modified content (this will overwrite the file)
        with open(file_path, 'w', encoding='utf-8') as file:
            for line in lines:
                # Split the line into components
                parts = line.split('\t')
                if len(parts) == 4:
                    # Replace space between indices with a tab
                    parts[2] = parts[2].replace(' ', '\t')
                    # Write the modified line
                    file.write('\t'.join(parts))
                else:
                    # If the line is malformed or doesn't match the expected format, just write it as is
                    file.write(line)

        print(f"Processed {filename}")

print("All files processed successfully!")
```
