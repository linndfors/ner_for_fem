```python
import os

def read_and_split_sentences(directory_path, df_par):
    counter = 0
    files_need_to_be_changed = []
    for filename in os.listdir(directory_path):
        wrong_file = 0
        ref_sent_ind_list = []
        if filename.endswith('.txt'):
            file_path = os.path.join(directory_path, filename)
        
            # Check if it's a file
            if os.path.isfile(file_path):
                
                with open(file_path, 'r', encoding='utf-8') as file:
                    content = file.read()
                    sentences = content.split('\n')
                    
                    # Filter out empty strings if needed
                    sentences = [s for s in sentences if s.strip()]
                    
                    for sentence in sentences:
                        df_par_exist = df_par[df_par['changed_sentence']==sentence]
                        if not df_par_exist.empty:
                            ref_sent_ind = df_par_exist['orig_sent_id'].values[0]
                            ref_sent_ind_list.append(ref_sent_ind)
                            # print(f"{ref_sent_ind} - {sentence}")
                        else:
                            wrong_file = 1
                            print("_______________THERE IS NO SUCH SENTENCE:", sentence)
                if wrong_file:
                    print(f"Reading file: {filename}")
                    counter += 1
                    files_need_to_be_changed.append(file_path)
                else:
                    meta_filename = filename.replace(".txt", ".meta")
                    meta_file_path = os.path.join(directory_path, meta_filename)
                    with open(meta_file_path, 'w', encoding='utf-8') as file:
                        for value in ref_sent_ind_list:
                            file.write(str(value) + '\n')

    return files_need_to_be_changed
```

### Rename files

```python
import os

def rename_filtered_files(directory_path):
    for filename in os.listdir(directory_path):
        if '_1_filtered' in filename:
            new_filename = filename.replace('_1_filtered', '-swapped')
            old_path = os.path.join(directory_path, filename)
            new_path = os.path.join(directory_path, new_filename)

            os.rename(old_path, new_path)
            print(f'Renamed: {filename} → {new_filename}')

# Example usage
directory = '/Users/linndfors/study/diploma/v2.0-swapped/bruk'
rename_filtered_files(directory)
```

```python
directory = '/Users/linndfors/study/diploma/v2.0-swapped/ng'
rename_filtered_files(directory)
```

### Create balanced dataset for NER

```python
def delete_meta_files(directory_path):
    for filename in os.listdir(directory_path):
        if filename.endswith('.meta'):
            file_path = os.path.join(directory_path, filename)
            if os.path.isfile(file_path):
                os.remove(file_path)
                print(f"Deleted: {filename}")
```

```python
directory = '/Users/linndfors/study/diploma/for_swapped_ner/ng_changed'
delete_meta_files(directory)
```

```python
directory = '/Users/linndfors/study/diploma/for_swapped_ner/bruk_changed'
delete_meta_files(directory)
```

```python
import os

def rename_swapped_files(directory_path):
    for filename in os.listdir(directory_path):
        if '-swapped' in filename:
            new_filename = filename.replace('-swapped', '_1')
            old_path = os.path.join(directory_path, filename)
            new_path = os.path.join(directory_path, new_filename)

            os.rename(old_path, new_path)
            print(f'Renamed: {filename} → {new_filename}')
```
